```python
postPath = "stackoverflow.com-Posts/Posts.xml"
```

```python
questions = []
```

For each post in Posts, we try to find the github referrence in Post body.

```python
with open(postPath, encoding='utf-8') as posts:
    for line in posts:
        if "<?xml" in line or "<posts>" in line:
            continue
        try:
            parsed = et.fromstring(line)
        except:
            continue
        if parsed.get("PostTypeId") != "1":
            continue
        id = int(parsed.get("Id"))
        if id % 100000 == 0:
            print(id)
            
        if "github.com/aws-samples" in parsed.get("Body") or "github.com/Azure-Samples" in parsed.get("Body") or "github.com/googlesamples" in parsed.get("Body") or "github.com/googlearchive" in parsed.get("Body") or "github.com/spring-guides" in parsed.get("Body") or "github.com/SAP-samples" in parsed.get("Body") or "github.com/spring-cloud-samples" in parsed.get("Body") or "github.com/spring-guides" in parsed.get("Body"):
            questions.append(parsed)
```

Path of code samples by framework

Filtring only questions about the selected code samples

```python
def filterSamples(samplesPath, questions):
    lines = []
    for samplePath in samplesPath["full_name"]:
        owner = samplePath.split("/")[0]
        sample = samplePath.split("/")[1]
        for question in questions:
            if sample in question.get("Body"):
                print(f"{question.get('Id')}")
                lines.append([owner, sample, question.get("Id"), question.get("CreationDate"), question.get("OwnerUserId"), question.get("PostTypeId"), question.get("AcceptedAnswerId")])
    return lines
```

```python
def createDataFrame(lines):
    columns = ["framework", "path", "id", "creationDate", "ownerUserId", "postTypeId", "acceptedAnswerId"]
    df = pd.DataFrame(lines, columns=columns)
    return df
```

Transforming data to DataFrame

```python
samplesPath = pd.read_csv("../0-GetCodeSamples/codesamples.csv")
```

```python
samplesDF = createDataFrame(filterSamples(samplesPath, questions))
samplesDF.to_csv("sampleQuestions.csv", index=False)
samplesDF["path"].drop_duplicates().to_csv("sampleWithQuestions.csv", index=False)
```
