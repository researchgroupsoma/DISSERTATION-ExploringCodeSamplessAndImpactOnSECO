```python
import pandas as pd
```

```python
questions = pd.read_csv("questions.csv")
```

```python
modify = questions[questions["Category"] == "MODIFY"]
```

```python
android = modify[modify["Framework"]== "Android"]
```

```python
spring = modify[modify["Framework"]== "Spring"]
```

```python
def count(tags):
  count = dict()
  for tag in tags:
    for t in tag.split("><"):
      t = t.replace("<", "")
      t = t.replace(">", "")
      #t = t.replace("android-", "") if "android-" in t else t
      if t in count.keys():
        count[t] = count[t] + 1
      else:
        count [t] = 1
  return count
```

```python
androidTags = count(android["Tags"])
```

```python
springTags = count(spring["Tags"])
```

```python
androidDF = pd.DataFrame.from_dict(androidTags.items()).sort_values(by=1, ascending=False)
```

```python
androidDF[androidDF[0] == "camera-flash"]
```

```python
springDF = pd.DataFrame.from_dict(springTags.items()).sort_values(by=1, ascending=False)
```

```python
androidDF.to_csv("android-tags.csv", index=False, header=["Tag", "Count"])
```

```python
springDF.to_csv("spring-tags.csv", index=False, header=["Tag", "Count"])
```
