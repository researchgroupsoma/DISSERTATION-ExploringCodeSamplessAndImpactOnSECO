```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
```

```python
df1 = pd.read_csv("android.csv", usecols=[20])
```

```python
df2 = pd.read_csv("spring.csv", usecols=[20])
```

```python
df1["Questions"] = 1
```

```python
df1
```

```python
df2["Questions"] = 1
```

```python
df1 = df1.groupby(by="Category").sum()
```

```python
df1
```

```python
df2 = df2.groupby(by="Category").sum()
```

```python
df1 = df1.sort_values("Questions")
```

```python
df2 = df2.sort_values("Questions")
```

```python
df1
```

```python
df1.to_csv("android_questions.csv")
```

```python
df2.to_csv("spring_questions.csv")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# create dummy pandas dataframes


#create subplot figure with having two side by side plots
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
print(axes)
# plot first pandas frame in subplot style
pie1 = df1.plot(ax = axes[0], kind="pie", y="Questions", colormap="tab20c", autopct="%1.1f%%", fontsize=20, figsize=(10,7), legend=False, labeldistance=None) 
pie1.set_title("Android Questions", fontsize=20)
pie1.set_ylabel("")
pie1.legend(labels=["Importing", "Running", "Reference", "Modification"], prop={"size":16}, bbox_to_anchor=(1.45,0.1))

# plot second pandas frame in subplot style
pie2 = df2.plot(ax = axes[1], subplots=False, kind="pie", y="Questions", colormap="tab20c", autopct="%1.1f%%", fontsize=20, figsize=(10,7), legend=False, labeldistance=None)
pie2.set_title("Spring Questions", fontsize=20)
pie2.set_ylabel("")

fig.savefig("questions_category_pie.pdf", bbox_inches="tight")
```
