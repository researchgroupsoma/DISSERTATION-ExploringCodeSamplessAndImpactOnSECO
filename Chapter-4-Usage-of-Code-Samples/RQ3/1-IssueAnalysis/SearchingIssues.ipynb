```
from github import Github
g = Github("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")
import pandas as pd
```

```
def verifyIfHasCommitOnEvents(events):
  for event in events:
      if (event.commit_id != None):
        return True
```

```
def verifyIfHasPullOrCommitOnComments(comments):
    for comment in comments:
      if ("/pull/" in comment.body or "/commit/" in comment.body):
        return True
```

```
def getIssues(framework, sample, issues):
  repo = g.get_repo(sample)
  for issue in repo.get_issues(state="closed"):
    try:
      if("issues" in issue.html_url and (verifyIfHasCommitOnEvents(issue.get_events()) or verifyIfHasPullOrCommitOnComments(issue.get_comments()))):
        print(issue.html_url)
        issues.append({
                                "framework": framework,
                                "sample": sample,
                                "issue_id": issue.id, 
                                "issue_number": issue.number,
                                "issue_title": issue.title,
                                "issue_state": issue.state,
                                "issue_html": issue.html_url
                                })
    except Exception as e:
      print(e)
  return issues
```

```
samples = pd.read_csv("googlesamples.txt", names=["sample"])
```

```
dataframe = pd.DataFrame(columns=["framework", "sample", "issue_id", "issue_number", "issue_title", "issue_state", "issue_html"])
```

```
framework = "Android"
issues = []
for sample in samples["sample"]:
  issues = getIssues(framework, sample, issues)
```

```
df = dataframe.append(other=issues)
```

```
df = df[df["sample"] != "googlesamples/android-BasicRenderScript"]
```
