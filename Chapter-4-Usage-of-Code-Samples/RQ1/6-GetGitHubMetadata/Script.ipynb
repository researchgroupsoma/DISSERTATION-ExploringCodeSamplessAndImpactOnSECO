```python
valid_tuples = pd.read_csv("../5-RepoFilter/output.csv", index_col=0)
```

```python
dataframe = pd.DataFrame()
cache = dict()
for index in valid_tuples.index:
    print((index/valid_tuples.index[-1])*100)
    repository = valid_tuples.loc[index]
    if (repository["watcher_repo"] in cache.keys()):
        print("Already assessed!")
        continue
    try:
        repo = g.get_repo(repository["watcher_repo"])
    except Exception as e:
        print(e)
    cache[repository["watcher_repo"]] = repo
    objeto = {
        "framework": repository["framework"],
        "sample": repository["sample"],
        "full_name": repo.full_name,
        "archived": repo.archived,
        "created_at": repo.created_at,
        "description": repo.description,
        "forks": repo.forks_count,
        "language": repo.language,
        "open_issues": repo.get_issues(state="open").totalCount,
        "closed_issues": repo.get_issues(state="closed").totalCount,
        "stars": repo.stargazers_count,
        "subscribers": repo.subscribers_count,
        "updated_at": repo.updated_at,
        "watchers": repo.watchers_count,
        "commits": repo.get_commits().totalCount,
        "downloads": repo.get_downloads().totalCount
    }
    objeto["development_time"] = (objeto["updated_at"] - objeto["created_at"]).days
    dataframe = dataframe.append(objeto, ignore_index=True)
```

```python
dataframe.to_csv("output.csv")
```
