```python
import pandas as pd
```

```python
authors = pd.read_csv("../1-GetMaintainersFromCommits/output_savepoint.csv", index_col=0)
```

```python
authors
```

```python
mergers = pd.read_csv("../2-GetMaintainersFromPR/output_savepoint.csv", index_col=0)
```

```python
mergers
```

```python
concat = pd.concat([authors, mergers], ignore_index=True)
```

```python
concat
```

```python
concat["framework"] = "Framework"
for index in concat.index:
    out = concat.loc[index]
    sample = out["sample"]
    if ("googlesamples" in sample or "googlearchive" in sample):
        concat.loc[index]["framework"] = "Android"
    elif ("spring-guides" in sample):
        concat.loc[index]["framework"] = "Spring"
    elif("Azure-Samples" in sample):
        concat.loc[index]["framework"] = "Azure"
    elif("aws-samples" in sample):
        concat.loc[index]["framework"] = "AWS"
```

```python
concat.drop("sample", 1, inplace=True)
```

```python
concat.drop_duplicates(inplace=True)
```

```python
concat.to_csv("output.csv")
```
