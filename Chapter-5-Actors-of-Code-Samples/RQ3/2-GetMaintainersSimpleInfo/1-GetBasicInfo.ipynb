```python
from github import Github
import pandas as pd
import traceback
```

```python
maintainers = pd.read_csv("../1-GetMaintainersList/3-Joint/output.csv", index_col=1)
```

```python
g = Github("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")
```

```python
del maintainers["Unnamed: 0"]
```

```python
new_maintainers = pd.DataFrame()
```

```python
for maintainer in maintainers.index:
    print(maintainer)
    try:
        user = g.get_user(maintainer)
        properties = user.__dict__["_rawData"]
        serie = maintainers.loc[maintainer].append(pd.Series(properties).rename(maintainer))
    except Exception as e:
        traceback.print_exc()
        print(maintainer)
        continue
    new_maintainers = new_maintainers.append(serie)
```

```python
def set_github_time():
    new_maintainers["created_at"] = pd.to_datetime(new_maintainers["created_at"])
    new_maintainers["updated_at"] = pd.to_datetime(new_maintainers["updated_at"])
    today = pd.to_datetime("2022-04-13", utc=True)
    new_maintainers["github_time"] = (today - new_maintainers['created_at']).dt.days
```

```python
set_github_time()
```

```python
new_maintainers.to_csv("output.csv", index=False)
```
