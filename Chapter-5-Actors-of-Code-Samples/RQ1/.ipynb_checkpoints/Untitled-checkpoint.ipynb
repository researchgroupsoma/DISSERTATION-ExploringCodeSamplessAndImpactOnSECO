```python
import fnmatch
import os
import pandas as pd
```

```python
def find_paths(pattern, path):
    result = []
    for root, dirs, files in os.walk(path):
        if '.git' in root:
            continue
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                result.append(os.path.join(root, name))
    return result
```

```python
paths = find_paths("*.csv", "C:\\Users\\Gabriel\\Documents\\gabrielsmenezes\\pesquisamestrado\\metricsbycommits\\azure-samples")
```

```python
usefull = []
for p in paths:
    with open(p) as f:
        lines = []
        for l in f:
            lines.append(l)
            print (l)
        usefull.append(lines[-1].replace("\n", ""))
```

```python
df = pd.DataFrame([u.split(",") for u in usefull], columns=["framework","path","commits","date","numberOfJavaFiles","countLineCode/numberOfJavaFiles","SumCyclomaticStrict/CountDeclMethod","readability"])
```

```python
df.sort_values(by="path", inplace=True)
```
